```python
import matplotlib.pyplot as plt
from scipy import interpolate
x= np.squeeze(np.array([[1],[0]]))
y= np.squeeze(np.array([[0],[1]]))
f = interpolate.interp1d(x, y)
print(x[0])
print(10)
xnew = np.linspace(x[0], x[1], 10)[1:-1]
print(xnew)
xnew_new_size = np.expand_dims(xnew,axis=1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
ynew_new_size = np.expand_dims(ynew,axis=1)
new = np.concatenate((xnew_new_size,ynew_new_size),axis=1)
print(new)
plt.plot(x, y, 'o', xnew, ynew, '.')
plt.show()
```
